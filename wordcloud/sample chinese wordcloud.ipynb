```python
from wordcloud import WordCloud 

text= open('chun.txt','r').read()

wordcloud = WordCloud(background_color="white",font_path=r"ziti.ttf",).generate(text)

import matplotlib.pyplot as plt
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import jieba
cut_text = ' '.join(jieba.cut(text))
wordcloud = WordCloud(background_color="white",font_path=r"ziti.ttf",).generate(cut_text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import imageio
pic = imageio.imread('a.png')
wordcloud = WordCloud(mask=pic,
    background_color="white",font_path=r"ziti.ttf",).generate(cut_text)

plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from wordcloud import ImageColorGenerator
image_color = ImageColorGenerator(pic)
a= wordcloud.recolor(color_func = image_color)
plt.imshow(a,interpolation='bilinear')
plt.axis("off")
plt.show()
```
