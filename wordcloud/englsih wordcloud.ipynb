```python
from wordcloud import WordCloud,ImageColorGenerator
import imageio
import matplotlib.pyplot as p
```

```python
text=open("sample.rtf",'r').read()
picture=imageio.imread("a.png")
wordcloud=WordCloud(background_color="white",scale=10,mask=picture).generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
wordcloud.to_file("e.png")
```
